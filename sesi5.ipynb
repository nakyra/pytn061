## Combining Data

```python
city_data
```

```python
city_data_2 = pd.DataFrame(
    {"revenue": [6547, 486], "employee": [14, 5]},
    index=["Kediri", "Sumba"]
)
city_data_2
```

```python
all_city = pd.concat([city_data, city_data_2])
all_city
```

```python
city_countries = pd.DataFrame({
    "country": ["Indonesia", "Japan", "Indonesia", "Canada", "Spain"],
    "capital": [1, 1, 0, 0, 0]},
    index=["Jakarta", "Tokyo", "Bali", "Toronto", "Barcelona"]
)
city_countries
```

```python
cities = pd.concat([all_city, city_countries], axis=1)
cities
```

```python
cities = pd.concat([all_city, city_countries], axis=1, join='inner')
cities
```

```python
cities = pd.concat([all_city, city_countries], axis=1, join='outer')
cities
```
